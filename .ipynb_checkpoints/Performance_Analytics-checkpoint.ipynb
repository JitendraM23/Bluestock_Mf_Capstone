```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

sns.set_theme(style="whitegrid")
DB_PATH = "bluestock_mf.db"
conn = sqlite3.connect(DB_PATH)
print("Connected to bluestock_mf.db for Performance Analytics")

query = """
SELECT n.date, n.amfi_code, f.scheme_name, n.nav
FROM fact_nav n
JOIN dim_fund f ON n.amfi_code = f.amfi_code
ORDER BY n.amfi_code, n.date;
"""
df_nav = pd.read_sql_query(query, conn)
df_nav['date'] = pd.to_datetime(df_nav['date'])

df_nav['daily_return'] = df_nav.groupby('amfi_code')['nav'].pct_change()
df_returns = df_nav.dropna(subset=['daily_return']).copy()
plt.figure(figsize=(12, 6))
sns.histplot(df_returns['daily_return'], bins=200, kde=True, color='indigo')

plt.title("Sanity Check: Distribution of Daily Returns Across All 40 Schemes", fontsize=14, fontweight='bold')
plt.xlabel("Daily Return (Expressed as a Decimal, e.g., 0.01 = 1%)", fontsize=12)
plt.ylabel("Frequency (Number of Trading Days)", fontsize=12)
plt.axvline(x=0, color='red', linestyle='--', label="Zero Return Line")
plt.legend()
import os
os.makedirs("charts", exist_ok=True)
plt.tight_layout()
plt.savefig("charts/validation_daily_returns.png", dpi=300)
plt.show()

print("\n Daily Returns Statistical Summary:")
print(df_returns['daily_return'].describe())
```

```python
metrics_df = pd.DataFrame(index=df_nav['amfi_code'].unique())
metrics_df.index.name = 'amfi_code'

scheme_map = df_nav.drop_duplicates('amfi_code').set_index('amfi_code')['scheme_name']
metrics_df.insert(0, 'scheme_name', metrics_df.index.map(scheme_map))
latest_date = df_nav['date'].max()
print(f"from the latest date: {latest_date.strftime('%Y-%m-%d')}\n")

for years, col_name in [(1, '1Yr_CAGR'), (3, '3Yr_CAGR'), (5, '5Yr_CAGR')]:
    target_start_date = latest_date - pd.DateOffset(years=years)
    
    cagr_list = []
    
    for amfi in metrics_df.index:
        fund_data = df_nav[df_nav['amfi_code'] == amfi]
        
        window = fund_data[fund_data['date'] >= target_start_date]
        
        if window.empty:
            cagr_list.append(np.nan)
            continue
            
        start_row = window.iloc[0]
        end_row = window.iloc[-1]
        actual_days = (end_row['date'] - start_row['date']).days
        if actual_days < (years * 365 * 0.8): 
            cagr_list.append(np.nan)
        else:
            n_years = actual_days / 365.25
            cagr = (end_row['nav'] / start_row['nav']) ** (1 / n_years) - 1
            cagr_list.append(cagr)
            
    metrics_df[col_name] = cagr_list

display_df = metrics_df.copy()
for col in ['1Yr_CAGR', '3Yr_CAGR', '5Yr_CAGR']:
    display_df[col] = (display_df[col] * 100).round(2).astype(str) + '%'

# 5. Show the top 10 rows to validate the outputs
print("CAGR Performance Matrix (Sample):")
display(display_df.head(10))
```

```python
rf_rate = 0.065  
annualized_return = df_returns.groupby('amfi_code')['daily_return'].mean() * 252
annualized_volatility = df_returns.groupby('amfi_code')['daily_return'].std() * np.sqrt(252)
sharpe_ratios = (annualized_return - rf_rate) / annualized_volatility
metrics_df['Sharpe_Ratio'] = sharpe_ratios.round(3)
metrics_df['Sharpe_Rank'] = metrics_df['Sharpe_Ratio'].rank(ascending=False).astype(int)
print("Top 5 Funds by Risk-Adjusted Performance (Sharpe Ratio):")
display_columns = ['scheme_name', 'Sharpe_Ratio', 'Sharpe_Rank']
display(metrics_df[display_columns].sort_values('Sharpe_Rank').head(5))
```

```python
negative_returns = df_returns[df_returns['daily_return'] < 0]
downside_volatility = negative_returns.groupby('amfi_code')['daily_return'].std() * np.sqrt(252)
sortino_ratios = (annualized_return - rf_rate) / downside_volatility
metrics_df['Sortino_Ratio'] = sortino_ratios.round(3)
metrics_df['Sortino_Rank'] = metrics_df['Sortino_Ratio'].rank(ascending=False).astype(int)
print("Top 5 Funds by Downside Protection (Sortino Ratio):")
display_columns = ['scheme_name', 'Sharpe_Ratio', 'Sortino_Ratio', 'Sortino_Rank']
display(metrics_df[display_columns].sort_values('Sortino_Rank').head(5))
```

```python
market_returns = df_returns.groupby('date')['daily_return'].mean().rename('market_return')
df_reg = df_returns.merge(market_returns, on='date', how='inner')
alpha_list = []
beta_list = []
for amfi in metrics_df.index:
    fund_data = df_reg[df_reg['amfi_code'] == amfi]
    
    if len(fund_data) < 252:
        alpha_list.append(np.nan)
        beta_list.append(np.nan)
        continue
        
    X = fund_data['market_return']
    Y = fund_data['daily_return']
    
    # Run the regression using Scipy!
    slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)
    
    beta_list.append(slope)
    annualized_alpha = intercept * 252
    alpha_list.append(annualized_alpha)
metrics_df['Beta'] = np.round(beta_list, 3)
metrics_df['Alpha_Pct'] = (np.array(alpha_list) * 100).round(2)

metrics_df['Alpha_Rank'] = metrics_df['Alpha_Pct'].rank(ascending=False).astype(int)
alpha_beta_export = metrics_df[['scheme_name', 'Alpha_Pct', 'Beta']]
alpha_beta_export.to_csv("alpha_beta.csv")
print("Successfully exported 'alpha_beta.csv' to your workspace!\n")print("Top 5 Funds by Pure Manager Skill (Alpha):")
display_columns = ['scheme_name', 'Alpha_Pct', 'Beta', 'Alpha_Rank']
display(metrics_df[display_columns].sort_values('Alpha_Rank').head(5))
```

```python
df_nav['running_max'] = df_nav.groupby('amfi_code')['nav'].cummax()
df_nav['drawdown'] = (df_nav['nav'] / df_nav['running_max']) - 1
max_drawdowns = df_nav.groupby('amfi_code')['drawdown'].min()
metrics_df['Max_Drawdown'] = max_drawdowns.round(4)
metrics_df['Max_DD_Rank'] = metrics_df['Max_Drawdown'].rank(ascending=False).astype(int)

top_fund_amfi = metrics_df['Max_DD_Rank'].idxmin()
top_fund_dd_data = df_nav[df_nav['amfi_code'] == top_fund_amfi]
worst_day = top_fund_dd_data.loc[top_fund_dd_data['drawdown'].idxmin()]
metrics_df['Max_DD_Pct'] = (metrics_df['Max_Drawdown'] * 100).round(2).astype(str) + '%'

print(f"Best Fund at surviving crashes is {metrics_df.loc[top_fund_amfi, 'scheme_name']}")
print(f"worst drop was {worst_day['drawdown']*100:.2f}%, which happened on {worst_day['date'].strftime('%Y-%m-%d')}\n")
print("🛡️ Top 5 Funds by Maximum Drawdown Protection:")
display_columns = ['scheme_name', 'Max_DD_Pct', 'Max_DD_Rank']
display(metrics_df[display_columns].sort_values('Max_DD_Rank').head(5))
```

```python
exp_query = "SELECT amfi_code, expense_ratio_pct FROM fact_performance;"
df_exp = pd.read_sql_query(exp_query, conn).set_index('amfi_code')
metrics_df = metrics_df.join(df_exp)
metrics_df['3Yr_Rank'] = metrics_df['3Yr_CAGR'].rank(ascending=False)
metrics_df['Exp_Rank'] = metrics_df['expense_ratio_pct'].rank(ascending=True)
def rank_to_score(rank_col):
    max_rank = metrics_df[rank_col].max()
    return ((max_rank - metrics_df[rank_col]) / (max_rank - 1)) * 100
metrics_df['Total_Score'] = (
    0.30 * rank_to_score('3Yr_Rank') +
    0.25 * rank_to_score('Sharpe_Rank') +
    0.20 * rank_to_score('Alpha_Rank') +
    0.15 * rank_to_score('Exp_Rank') +
    0.10 * rank_to_score('Max_DD_Rank')
)
scorecard_df = metrics_df.dropna(subset=['Total_Score']).sort_values('Total_Score', ascending=False)
scorecard_df['Total_Score'] = scorecard_df['Total_Score'].round(2)
scorecard_export = scorecard_df[['scheme_name', 'Total_Score', '3Yr_CAGR', 'Sharpe_Ratio', 'Alpha_Pct', 'expense_ratio_pct', 'Max_Drawdown']]
scorecard_export.to_csv('fund_scorecard.csv')
print("Successfully exported 'fund_scorecard.csv' to your workspace!\n")
print("THE ULTIMATE FUND SCORECARD (Top 10):")
display_columns = ['scheme_name', 'Total_Score', '3Yr_Rank', 'Sharpe_Rank', 'Alpha_Rank']
display(scorecard_df[display_columns].head(10))
```

```python
top_5_amfis = scorecard_df.head(5).index.tolist()
start_date_3yr = df_returns['date'].max() - pd.DateOffset(years=3)
df_3yr = df_returns[df_returns['date'] >= start_date_3yr].copy()
market_3yr = market_returns[market_returns.index >= start_date_3yr].copy()

plt.figure(figsize=(14, 7))
print("Calculating Tracking Errors for Top 5 Funds vs Benchmark:")
for amfi in top_5_amfis:
    fund_data = df_3yr[df_3yr['amfi_code'] == amfi].set_index('date')
    scheme_name = scorecard_df.loc[amfi, 'scheme_name']
    
    fund_data['cumulative_return'] = (1 + fund_data['daily_return']).cumprod() - 1

    aligned_data = pd.concat([fund_data['daily_return'], market_3yr], axis=1).dropna()
    active_return = aligned_data['daily_return'] - aligned_data['market_return']
    tracking_error = active_return.std() * np.sqrt(252)
    
    print(f"- {scheme_name}: Tracking Error = {(tracking_error * 100):.2f}%")
    
    plt.plot(fund_data.index, fund_data['cumulative_return'] * 100, label=f"{scheme_name[:25]}...", linewidth=2)

market_cumulative = (1 + market_3yr).cumprod() - 1
plt.plot(market_3yr.index, market_cumulative * 100, label="Benchmark Proxy (Nifty 100)", 
         color='black', linewidth=3, linestyle='--')

plt.title("Top 5 Scorecard Funds vs Market Benchmark (3-Year Cumulative Returns)", fontsize=15, fontweight='bold', pad=15)
plt.xlabel("Timeline", fontsize=12)
plt.ylabel("Cumulative Growth Return (%)", fontsize=12)
plt.legend(loc='upper left', frameon=True, shadow=True)
plt.tight_layout()
plt.savefig("charts/benchmark_comparison_chart.png", dpi=300)
plt.show()
print("\n Successfully exported 'benchmark_comparison_chart.png' to your charts folder")
```
